```python
#Ejercicio 2

def process_text(texto="Texto de ejemplo   ", cad=["texto", "ejemplo"]):
    texto=texto.lower()
    texto=texto.strip()
    contador=0

    for palabra in cad:
        if palabra in texto:
            texto=texto.replace(palabra, "*"*len(palabra) )
            contador=contador+1
    return texto, contador

def main():
    cadena, reemplazos = process_text()
    print("\nCadena procesada: ", cadena)
    print("Palabras reemplazadas: ", reemplazos)

if __name__ == "__main__":
    main()


```

```python
#Ejercicio 3

def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5)+1):
        if num%i==0:
            return False
    return True

def prime_list(lim):
    num_primo=[]
    for num in range(2, lim+1):
        if is_prime(num):
            num_primo.append(num)
    return num_primo

def check_palindrome(primos):
    palindromos=[]
    for num in primos:
        if str(num)==str(num)[::-1]:
            palindromos.append(num)
    return palindromos

def categorize_prime(primo):
    if primo<10:
        return "pequeño"
    elif primo>=10 and primo<=99:
        return "mediano"
    else:
        return "grande"

def main():
    n=int(input("\nIntroduce un numero: "))
    if is_prime(n):
        print("Es un numero primo\n")
    else:
        print("No es un numero primo\n")

    limite=int(input("Pon un límite de números: "))
    primos=prime_list(limite)
    print("\nLos numeros primos hasta ", limite," :", primos)

    palindromos=check_palindrome(primos)
    print("\nLos numeros palindromos son: ", palindromos)

    for primo in primos:
        categoria=categorize_prime(primo)
        print("\nEl numero primo ", primo, " es categoria: ", categoria)

if __name__ == "__main__":
    main()
```

```python
#Ejercicio 4

def suma(num):
    return sum(num)

def suma_dif(numeros):
    sum_neg=sum(i for i in numeros if i < 0)
    sum_pos=sum(i for i in numeros if i > 0)
    return sum_neg, sum_pos

def maxmin(numeros):
    return max(numeros), min(numeros)

def unicos(numeros):
    return len(numeros)==len(set(numeros))

def pos_neg(numeros):
    cont_neg=0
    cont_pos=0
    for i in numeros:
        if i<0:
            cont_neg+=1
        else:
            cont_pos+=1
    if cont_pos<cont_neg:
        return "La lista contiene mas numeros negativos"
    elif cont_pos>cont_neg:
        return "La lista contiene mas numeros positivos"
    else:
        return "La lista contiene la misma cantidad de numeros positivos y negativos"

def main():

    numeros = list(map(int, input("Introduce una lista de números enteros separados por espacios: ").split()))

    while True:
        print("\nMenú de opciones:")
        print("1. Sumar todos los números.")
        print("2. Sumar los números positivos y negativos por separado.")
        print("3. Encontrar el número máximo y mínimo.")
        print("4. Comprobar si todos los números son únicos.")
        print("5. Ver si hay más números positivos o negativos.")
        print("6. Salir.")

        opcion = input("\nElige una opción (1-6): ")

        if opcion == '1':
            total = sum(numeros)
            print("La suma de todos los números es: ", total)

        elif opcion == '2':
            suma_negativos, suma_positivos = suma_dif(numeros)
            print("Suma de los números positivos: ", suma_positivos)
            print("Suma de los números negativos: ", suma_negativos)

        elif opcion == '3':
            maximo, minimo = maxmin(numeros)
            print("El número máximo es: ", maximo)
            print("El número mínimo es: ", minimo)

        elif opcion == '4':
            if unicos(numeros):
                print("Todos los números de la lista son únicos.")
            else:
                print("La lista contiene números repetidos.")

        elif opcion == '5':
            mensaje = pos_neg(numeros)
            print(mensaje)

        elif opcion == '6':
            print("Saliendo del programa. ¡Hasta luego!")
            break

        else:
            print("Opción no válida, por favor intenta de nuevo.")


if __name__ == "__main__":
    main()
```

```python
#Ejercicio 5

tareas=[]

def add_task(tarea):
    tareas.append(tarea)
    print("Tarea agregada.\n")

def remove_task(tarea):
    if tarea in tareas:
        tareas.remove(tarea)
        print("Tarea eliminada.\n")
    else:
        print("No se encontro la tarea a eliminar.\n")


def list_tasks():
    if not tareas:
        print("No existen tareas pendientes actualmente.\n")
    else:
        print("Tareas pendientes: ")
        for index, tarea in enumerate(tareas, start=1):
            print(f"{index}. {tareas}")


def task_manager():

    def modify_task():
        if tareas:
            nueva_tarea=input("Introduce la nueva tarea: ")
            print("Tarea modificada.\n")
            tareas[0]=nueva_tarea
        else:
            print("No existen tareas para modificar.\n")

    while True:
        print("\nGestor de Tareas")
        print("1. Agregar tarea")
        print("2. Eliminar tarea")
        print("3. Lista de tareas")
        print("4. Modificar primera tarea")
        print("5. Salir")

        opcion = input("Elige una opción (1-5): ")

        if opcion == '1':
            tarea = input("Introduce la tarea que deseas agregar: ")
            add_task(tarea)
        elif opcion == '2':
            tarea = input("Introduce la tarea que deseas eliminar: ")
            remove_task(tarea)
        elif opcion == '3':
            list_tasks()
        elif opcion == '4':
            modify_task()
        elif opcion == '5':
            print("Saliendo del gestor de tareas.")
            break
        else:
            print("Opción no válida. Por favor, elige otra opción.")


def main():
    task_manager()


if __name__ == '__main__':
    main()
```

```python
#Ejercicio 6

def imprime_cadena(cad, letra):
    for cadena in cad:
        if cad.startswith(str(letra)):
            print(cadena)


def contar(cad, subcad):
    contador=0
    for cadena in cad:
        if subcad in cad:
            contador=contador+1
    return contador


def larga_corta(cad):
    if not cad:
        return None, None
    cad_larga=max(cad, key=len)
    cad_corta=min(cad, key=len)
    return cad_larga, cad_corta


def igual(cad1, cad2):
    return cad1 is cad2


def mayor10(cad):
    for cadena in cad:
        if len(cad)>10:
            print("La cadena tiene tiene mas de 10 caracteres.\n")
            break
    else:
        print("No existe ninguna cadena que supere los 10 caracteres.\n")


def main():
    cadenas=[]
    num=int(input("Introduce cuantas cadenas quieres introducir: "))
    for _ in range(num):
        cadena=input("Introduce una cadena: ")
        cadenas.append(cadena)

    while True:
        print("\nMenú de Operaciones:")
        print("1. Imprimir cadenas que comienzan con una letra específica.")
        print("2. Contar cuántas cadenas contienen una subcadena.")
        print("3. Encontrar la cadena más larga y la más corta.")
        print("4. Verificar si dos cadenas son el mismo objeto.")
        print("5. Verificar si alguna cadena tiene más de 10 caracteres.")
        print("6. Salir")

        opcion = input("Elige una opción (1-6): ")

        if opcion == '1':
            letra = input("Introduce una letra: ")
            imprime_cadena(cadenas, letra)
        elif opcion == '2':
            subcadena = input("Introduce una subcadena: ")
            contador = contar(cadenas, subcadena)
            print("Hay ", contador, " cadenas que contienen la subcadena", subcadena)
        elif opcion == '3':
            larga, corta = larga_corta(cadenas)
            if larga and corta:
                print("La cadena más larga es: ", larga)
                print("La cadena más corta es: ", corta)
            else:
                print("No hay cadenas en la lista.")
        elif opcion == '4':
            cadena1 = input("Introduce la primera cadena: ")
            cadena2 = input("Introduce la segunda cadena: ")
            if cadena1 in cadenas and cadena2 in cadenas:
                son_mismos = igual(cadenas, cadena1, cadena2)
                if son_mismos:
                    print("Las cadenas son el mismo objeto.")
                else:
                    print("Las cadenas no son el mismo objeto.")
            else:
                print("Una o ambas cadenas no están en la lista.")
        elif opcion == '5':
            mayor10(cadenas)
        elif opcion == '6':
            print("Saliendo del programa.")
            break
        else:
            print("Opción no válida. Por favor, elige otra opción.")

if __name__ == '__main__':
    main()
```
